```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#PERTURBING INFECTION RATE

############################################################################################
#STEP 1: definition of the dynamical system 

def SIRD(z, t, b, m, d):
    S = z[0] # Susceptible compartment
    I = z[1] # Infectious compartment
    R = z[2] # Recovered compartment
    D = z[3] # Deceased compartment
    
    dSdt = -b * S * (I/N)
    dIdt =  b * S * (I/N) - m * I - d * I
    dRdt =                  m * I
    dDdt =                          d * I
    
    return [dSdt, dIdt, dRdt, dDdt]
############################################################################################



############################################################################################
#STEP 2: Initial conditions
z0 = [990, 10, 0, 0] # starting off with 990 healthy  and 10 infected people.
N = z0[0] + z0[1] + z0[2] # oi deceased de logizontai sto overall population.
############################################################################################



############################################################################################
#STEP 3: assign values to fixed parameters
m = 0.2 # m persons per day recover
d = 0.005
############################################################################################


############################################################################################
#STEP 4: define time-points
t = np.linspace(0,60)
############################################################################################



############################################################################################
#STEP 5: solve the model for different values of the quantity of interest and plot results

for R0 in np.arange(1, 7, 1): # Reproduction number


#solve SIRD model (solver: odeint)
    b = R0 * m # calculate infection rate b based on reproduction number R0
    
    z = odeint(SIRD, z0, t, args = (b, m, d))
    S = z[:,0]
    I = z[:,1]
    R = z[:,2]
    D = z[:,3]
    plt.plot(t, I, '--', linewidth = 2, label = 'R0 =' + str(R0))
############################################################################################




############################################################################################
#plot results
#plt.plot(t, S, 'r-', linewidth = 2, label = 'Susceptible Compartment')
#plt.plot(t, I, 'b--', linewidth = 2, label = 'Infectious Compartment')
#plt.plot(t, R, 'g:', linewidth = 2, label = 'Recovered Compartment')
#plt.plot(t, D, 'co', linewidth = 2, label = 'Deceased Compartment')

plt.xlabel('time (days)')
plt.ylabel('Infectious Compartment')
plt.legend()
plt.show()
############################################################################################
```
